```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def merge_names(df_names, df_prof, cols_name):
    df_result = pd.merge(df_names, df_prof, on='nconst', how='inner')
    df_result = df_result.drop('tconst', axis=1)
    df_result = df_result.drop_duplicates()
    df_result = df_result.rename(columns={"primaryName": cols_name})
    return df_result
```

```python
df = pd.read_csv('datasets/basics_knownForTitles_ratings.csv', index_col=0)
(df.tconst.value_counts() == 1).all()
df = df.query('numVotes >= 375')
df = df.dropna()
df = df.query('isAdult == 0')
df = df.drop(columns=['tconst', 'primaryTitle','isAdult'], axis=1)
df['nconst'] = df['nconst'].str.replace(',', ' ')
df['genres'] = df['genres'].str.replace(',', ' ')
df
```

```python
vec = CountVectorizer()
```

```python
genres = vec.fit_transform(df['genres']).toarray()
```

```python
df_genres = pd.DataFrame(data=genres, columns=vec.get_feature_names_out())
df_genres.columns
```

```python
df_producers = pd.read_csv('datasets/producers.csv')
df_producers_names = merge_names(df_names, df_producers, "producer")
```
